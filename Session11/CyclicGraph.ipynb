```python
import matplotlib.pyplot as plt
import math
def cyclicgraph(num_iter,ymin,ymax,ss):
    xp = []
    yp = []
    for i in range(num_iter):
        xp.append(i)
        cycle = math.floor(1+ i/(2*ss))
        x = abs(i/ss - 2*(cycle)+1)
    
        y = ymin + (ymax-ymin)*(1-x)
        yp.append(y)
    plt.plot(xp, yp)
    plt.axis("off")
    plt.show()
```

```python
cyclicgraph(num_iter = 1200, ymin = 20,ymax = 25,ss=200)
```
